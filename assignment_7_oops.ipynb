## Problem 1: Smart Light Controller

```python
# Class to control a smart light
class SmartLight:
    def __init__(self, name):
        # Initialize light with a name
        self.name = name
        self.status = "OFF"  # Default status is OFF
    
    def turn_on(self):
        # Turn the light ON
        self.status = "ON"
    
    def turn_off(self):
        # Turn the light OFF
        self.status = "OFF"
    
    def display_status(self):
        # Display current status of the light
        print(f"{self.name} is {self.status}")

# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

## Problem 2: Employee ID Card System

```python
# Class to manage employee ID card details
class Employee:
    def __init__(self, name, emp_id, department):
        # Initialize employee details
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    def display_id_card(self):
        # Display employee ID card information
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Sample Input
employee = Employee("Rahul", "EMP102", "IT")
employee.display_id_card()
```

## Problem 3: Mobile Contact Record

```python
# Class to store mobile contact information
class Contact:
    def __init__(self, name, phone_number):
        # Initialize contact details
        self.name = name
        self.phone_number = phone_number
    
    def display_contact(self):
        # Display saved contact information
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone_number}")

# Sample Input
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

## Problem 4: Product Price Tag Generator

```python
# Class to generate product price tags
class Product:
    def __init__(self, name, price):
        # Initialize product details
        self.name = name
        self.price = price
    
    def display_price_tag(self):
        # Display formatted price tag
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# Sample Input
product = Product("Headphones", 2499)
product.display_price_tag()
```

## Problem 5: Movie Rating Display System

```python
# Class to store and display movie ratings
class Movie:
    def __init__(self, name, rating):
        # Initialize movie details
        self.name = name
        self.rating = rating
    
    def display_details(self):
        # Display movie name and rating
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Sample Input
movie = Movie("Inception", 4.8)
movie.display_details()
```

## Problem 6: Delivery Address Manager

```python
# Class to manage delivery address details
class Delivery:
    def __init__(self, customer_name, address):
        # Initialize delivery details
        self.customer_name = customer_name
        self.address = address
    
    def display_delivery_details(self):
        # Display delivery information
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Sample Input
delivery = Delivery("Suman", "Hyderabad")
delivery.display_delivery_details()
```
